## Encoding Categorical Variables

Gender is encoded using label encoding .
This helps us check if adding demographic details makes the customer groups clearer.

```python
df_orginal  = df.copy()
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df.drop('CustomerID',axis=1,inplace=True)
df.head()
```

## Why Scaling is Required

K-Means relies on Euclidean distance.
Features with larger numeric ranges dominate distance calculations.

Example:
- Annual Income ranges up to ~140
- Spending Score ranges up to 100
- Age ranges up to ~70

Without scaling, income would disproportionately influence cluster formation.

StandardScaler centers data around a mean of 0 with a standard deviation of 1. It preserves the relationships between variables while ensuring no single feature dominates.

```python
scaler = StandardScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

### Feature Set 1: Income + Spending Score

This combination focuses purely on economic capacity and spending behavior.

```python
features_1 = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

### Feature Set 2: Age + Income + Spending Score

This set adds life-stage information to capture age-based spending patterns.

```python
features_2 = df[["Age", "Annual Income (k$)", "Spending Score (1-100)"]]
```

## Feature Set Selection Decision

After comparing different feature combinations:

- Feature Set 1 (Income + Spending) provides clear, interpretable clusters
- Feature Set 2 captures additional age-based variation but increases complexity

Final Choice:
Primary clustering will be performed using **Income and Spending Score**
Secondary analysis will use **Age + Income + Spending Score** for validation.
